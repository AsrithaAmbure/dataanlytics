```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("gym_members_exercise_tracking_synthetic_data.csv")

avg_workouthrs=df['workout_hrs'].mean()
print("Average workout hours:",avg_workouthrs)

max_calories=df['Calories_Burned'].max()
print("Maximum calories burned:",max_calories)

min_calories=df['Calories_Burned'].min()
print("Minimum calories burned:",min_calories)

index=df['Calories_Burned'].idxmin()
print("least active mem",index)

index=df['Calories_Burned'].idxmax()
print("least active mem",index)

plt.bar(df['workout_hrs'], df['Calories_Burned'], color='blue')
plt.xlabel('Workout Hours')
plt.ylabel('Calories Burned')
plt.title('Calories Burned vs. Workout Hours')
plt.show()

plt.scatter(df['Age'], df['Calories_Burned'], alpha=0.6, color='purple')
plt.xlabel('Age')
plt.ylabel('Calories Burned')
plt.title('Calories Burned vs. Age')
plt.show()

plt.hist(df['Calories_Burned'], bins=10, color='green')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.title('Distribution of Calories Burned')
plt.show()
```
