```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("weather_data (2).csv")

print(df)

avg = df["Temperature_C"].mean()
print("Average Temperature =", avg)

increase = df["Temperature_C"].diff().fillna(0)

print("Temperature Increased =")
print(increase.to_string(index=False))

sum_increase = increase.sum()
print("Sum of Increased Temperature =", sum_increase)

last_value = df["Temperature_C"].iloc[-1]

prediction = last_value + sum_increase

print("Predicted Future Temperature =", prediction)

if prediction > avg:
    print("Weather Temperature Will Increase")
else:
    print("Weather Temperature Stable")

plt.plot(df["Days"], df["Temperature_C"], marker='o')

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather Temperature Prediction")

plt.show()
```
